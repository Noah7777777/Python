# 소상공인 지역별 실적, 전망

### 1. 파일 전처리 저장 함수

```python
import pandas as pd
import numpy as np
import seaborn as sns
from openpyxl import load_workbook
import matplotlib.pyplot as plt
plt.rc('font', family='NanumGothic') # For Windows # 한글폰트

def small_business():
    # 파일 로드
    data = pd.read_excel("data/3. 소상공인_지역별_실적_및_전망_20200619152136.xlsx")
    # 열과 행 바꾸기
    data = data.T
    # 체감 전망 열 지우기
    del data[0]
    # 열 이름 바꾸기
    data.columns = ["서울","부산","대구", "인천","광주", "대전","울산","세종","경기","강원","충북","충남", "전북","전남","경북","경남","제주"]
    data.drop(['지역별(1)'], inplace=True)
    #마지막월은 체감이 없어서 마지막 월 전망의 순서를 조정
    data.loc['마지막월전망'] = data.iloc[-1]
    data.iloc[-2] = np.nan
    data.to_csv('C:/Users/acorn/Acon_jaeseo/teamproject/data/small_business_pre_data.csv', encoding='utf-8-sig')
    print("저장이 완료되었습니다.")
    
small_business()
```

### 2-1. 지역별 그래프

```python
### 전국 한번에 저장

import pandas as pd
import numpy as np
import seaborn as sns
from openpyxl import load_workbook
import matplotlib.pyplot as plt
plt.rc('font', family='NanumGothic') # For Windows # 한글폰트

def small_business():
    loc_list = ['서울','부산','대구','인천','광주','대전','울산','세종','경기','강원','충북','충남','전북','전남','경북','경남','제주']
    for loc_name in loc_list:
        data = pd.read_csv("C:/Users/acorn/Acon_jaeseo/teamproject/data/small_business_pre_data.csv", index_col='Unnamed: 0')
        plt.figure(figsize=(7,5))
        x1 = data[loc_name][0::2]  # 체감
        x2 = data[loc_name][1::2] # 전망

        # 체감의 평균치
        X1 = data.iloc[0:,0:][0::2] # 전체 지역 체감
        m1 = X1.mean(axis=1) # 체감 평균

        # # 전망의 평균치
        X2 = data.iloc[0:,0:][1::2] # 전체 지역 전망
        m2 = X2.mean(axis=1) # 전망 평균

        x2.plot(label = loc_name + " 전망")
        x1.plot(label = loc_name + " 체감")
        m2.plot.bar(label = "전망 평균", alpha=0.3)
        m1.plot.bar(label = "체감 평균", alpha=0.2, color="r")
        plt.xlabel("month")
        plt.ylabel("BSI")
        plt.title(loc_name+ " 소상공인 전망")
        plt.xticks(rotation=45)
        plt.legend(loc='lower left')
    #     plt.legend(bbox_to_anchor=(1.02, 1), borderaxespad=1)
        plt.grid(True, axis='y', color='gray', alpha=0.5, linestyle='--')
        plt.savefig('코로나_{}_소상공인전망.png'.format(loc_name)) # png형식 저장

small_business()
```
